```python
import pandas as pd
import numpy as np
```

```python
num_of_evaluators = 2
num_of_criteria = 15
criteria = ['C11','C21','C22','C23', 'C31', 'C32', 'C33', 'C34', 'C41', 'C42', 'C51', 'C52', 'C53',' C61', 'C62']
num_of_candidates = 10
candidates = ['Candidate '+str(i+1) for i in range(num_of_candidates)]
```

```python
path_first_eval = '../g-edas-matrix-results-v.xlsx'
results_eval_frst_df = pd.read_excel(path_first_eval, header=0)
results_eval_frst_df = results_eval_frst_df.fillna(0)
results_eval_frst_df = results_eval_frst_df.drop(results_eval_frst_df.columns[[0]], axis=1)
results_eval_frst_df
```

```python
path_second_eval = '../g-edas-matrix-results-s.xlsx'
results_eval_scnd_df = pd.read_excel(path_second_eval, header=0)
results_eval_scnd_df = results_eval_scnd_df.fillna(0)
results_eval_scnd_df = results_eval_scnd_df.drop(results_eval_scnd_df.columns[[0]], axis=1)
results_eval_scnd_df
```

```python
ling_scale = ['Veoma lose', 'Lose', 'Srednje lose', 'Prosecno', 'Srednje dobro', 'Dobro', 'Veoma dobro']
grey_nums = [[0, 1], [1, 3], [3, 4], [4, 5], [5, 6], [6, 9], [9, 10]]
scale = {'Linguistic Scale': ling_scale, 'Value': [i for i in range(7)], 'Grey Numbers': grey_nums}
scale_df = pd.DataFrame(data=scale)
scale_df
```

```python
def print_grey_num(grey_num):
    return '[ ' + str(grey_num[0]) + ', ' + str(grey_num[1]) + ' ]'
                
def print_matrix(matrix):
    matrix_dim = matrix.shape[0]
    for i in range(matrix_dim):
        for j in range(matrix_dim):
            print(print_grey_num(matrix[i,j,:]), end =" ")
        print("\n")
        
def get_str_matrix_of_grey_nums(matrix):
    str_matrix = [[] for i in range(matrix.shape[0])]
    
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            grey_num = matrix[i,j]
            str_matrix[i].append(print_grey_num(grey_num))
            
    return str_matrix
```

```python
def get_grey_num_for_value(df, row_ind, col_ind):
    value = df.iloc[row_ind,col_ind]
    grey_num_for_value = grey_nums[value]
    return np.array(grey_num_for_value)
```

```python
# replace values with grey nums; get g-edas matrix

edas_matrix = np.zeros((num_of_candidates, num_of_criteria, 2), dtype=float)

for i in range(num_of_candidates):
    for j in range(num_of_criteria):
        frst_grey_num = get_grey_num_for_value(results_eval_frst_df, i, j)
        scnd_grey_num = get_grey_num_for_value(results_eval_scnd_df, i, j)
        edas_matrix[i,j] = (frst_grey_num + scnd_grey_num) / 2
        
edas_matrix = np.around(edas_matrix, 3)
```

```python
# initial decision-making matrix

str_matrix = get_str_matrix_of_grey_nums(edas_matrix)
str_matrix_df = pd.DataFrame(str_matrix, columns = criteria)
str_matrix_df
```

```python
# grey average solution according to all criteria
average_solution_for_criteria_lower_bound = []
average_solution_for_criteria_upper_bound = []

for c in range(num_of_criteria):
    lower_bounds_for_c = np.ravel(edas_matrix[:,c,0])
    avg_lower_bound_for_c = sum(lower_bounds_for_c) / num_of_candidates
    average_solution_for_criteria_lower_bound.append(round(avg_lower_bound_for_c, 3))
    
    upper_bounds_for_c = np.ravel(edas_matrix[:,c,1])
    avg_upper_bound_for_c = sum(upper_bounds_for_c) / num_of_candidates
    average_solution_for_criteria_upper_bound.append(round(avg_upper_bound_for_c))
```

```python
avg_solution_for_criteria = {'Criteria': criteria, 'Lower bound of avg solution': average_solution_for_criteria_lower_bound, 'Upper bound of avg solution': average_solution_for_criteria_upper_bound}
avg_solution_for_criteria_df = pd.DataFrame(data=avg_solution_for_criteria)
avg_solution_for_criteria_df
```

```python
def get_evaluation_of_candidate_according_to_criteria(candidate, criterion):
    return edas_matrix[candidate,criterion]  

def calculate_pda_for_evaluation(candidate, criterion):
    grey_num_evaluation = get_evaluation_of_candidate_according_to_criteria(candidate, criterion)
    eval_l, eval_u = grey_num_evaluation[0], grey_num_evaluation[1]
    avg_sol_for_c_l = average_solution_for_criteria_lower_bound[criterion]
    avg_sol_for_c_u = average_solution_for_criteria_upper_bound[criterion]
    
    lower_pda_bound = max(0, (eval_l - avg_sol_for_c_u)) / 0.5*(avg_sol_for_c_l + avg_sol_for_c_u)
    upper_pda_bound = max(0, (eval_u - avg_sol_for_c_l)) / 0.5*(avg_sol_for_c_l + avg_sol_for_c_u)
    
    pda = np.array([lower_pda_bound, upper_pda_bound])
    return np.around(pda, 3)
    
    
def calculate_nda_for_evaluation(candidate, criterion):
    grey_num_evaluation = get_evaluation_of_candidate_according_to_criteria(candidate, criterion)
    eval_l, eval_u = grey_num_evaluation[0], grey_num_evaluation[1]
    avg_sol_for_c_l = average_solution_for_criteria_lower_bound[criterion]
    avg_sol_for_c_u = average_solution_for_criteria_upper_bound[criterion]
    
    lower_nda_bound = max(0, (avg_sol_for_c_l - eval_u)) / 0.5*(avg_sol_for_c_l + avg_sol_for_c_u)
    upper_nda_bound = max(0, (avg_sol_for_c_u - eval_l)) / 0.5*(avg_sol_for_c_l + avg_sol_for_c_u)
    
    nda = np.array([lower_nda_bound, upper_nda_bound])
    return np.around(nda, 3)
```

```python
# grey positive and negative distance from avg: PDA/NDA

pda_matrix = np.zeros((num_of_candidates, num_of_criteria, 2), dtype=float)
nda_matrix = np.zeros((num_of_candidates, num_of_criteria, 2), dtype=float)

for candidate in range(num_of_candidates):
    for criterion in range(num_of_criteria):
        pda_matrix[candidate,criterion] = calculate_pda_for_evaluation(candidate, criterion)
        nda_matrix[candidate,criterion] = calculate_nda_for_evaluation(candidate, criterion)
```

```python
# the positive grey distance from the average

pda_matrix_str = get_str_matrix_of_grey_nums(pda_matrix)
pda_matrix_df = pd.DataFrame(pda_matrix_str, columns = criteria)
pda_matrix_df
```

```python
# the negative grey distance from the average

nda_matrix_str = get_str_matrix_of_grey_nums(nda_matrix)
nda_matrix_df = pd.DataFrame(nda_matrix_str, columns = criteria)
nda_matrix_df
```

```python
# weights for criteria categories obtained through g-aph

weights = [0.259, 0.226, 0.107, 0.095, 0.108, 0.205] 
criteria_weights_from_g_aph = [weights[0]]
criteria_weights_from_g_aph.extend([weights[1] for i in range(3)]) # C21, C22, C23
criteria_weights_from_g_aph.extend([weights[2] for i in range(4)]) # C31, C32, C33, C34
criteria_weights_from_g_aph.extend([weights[3] for i in range(2)]) # C41, C42
criteria_weights_from_g_aph.extend([weights[4] for i in range(3)]) # C51, C52, C53
criteria_weights_from_g_aph.extend([weights[5] for i in range(2)]) # C61, C62
print(criteria_weights_from_g_aph)
```

```python
def get_weighted_dist_from_avg_for_candidate(candidate):
    pda_for_all_criteria = pda_matrix[candidate,:,:]
    nda_for_all_criteria = nda_matrix[candidate,:,:]
    weighted_pda = np.array([0., 0.])
    weighted_nda = np.array([0., 0.])
    
    for i in range(num_of_criteria):
        criterion_pda = pda_for_all_criteria[i]
        criterion_nda = nda_for_all_criteria[i]
        criterion_weight = criteria_weights_from_g_aph[i]
        
        weighted_pda += criterion_pda * criterion_weight
        weighted_nda += criterion_nda * criterion_weight
        
    return weighted_pda, weighted_nda

```

```python
# weighted sum of the grey PDA and NDA

weighted_pda_matrix = np.zeros((1, num_of_candidates, 2), dtype=float)
weighted_nda_matrix = np.zeros((1, num_of_candidates, 2), dtype=float)

for candidate in range(num_of_candidates):
    weighted_pda, weighted_nda = get_weighted_dist_from_avg_for_candidate(candidate)
    weighted_pda_matrix[0, candidate] = weighted_pda
    weighted_nda_matrix[0, candidate] = weighted_nda
    
weighted_pda_matrix = np.around(weighted_pda_matrix, 3)
weighted_nda_matrix = np.around(weighted_nda_matrix, 3)
```

```python
# weighted sum of the grey PDA

weighted_pda_str = get_str_matrix_of_grey_nums(weighted_pda_matrix)
weighted_pda_df = pd.DataFrame(weighted_pda_str, columns = ['Candidate '+str(i+1) for i in range(num_of_candidates)])
weighted_pda_df
```

```python
# weighted sum of the grey NDA

weighted_nda_str = get_str_matrix_of_grey_nums(weighted_nda_matrix)
weighted_nda_df = pd.DataFrame(weighted_nda_str, columns = ['Candidate '+str(i+1) for i in range(num_of_candidates)])
weighted_nda_df
```

```python
# normalizing values

pda_norm_factor = max(np.ravel(weighted_pda_matrix[:,:,1]))
nda_norm_factor = max(np.ravel(weighted_nda_matrix[:,:,1]))

norm_weighted_pda = np.around(weighted_pda_matrix / pda_norm_factor, 3)
norm_weighted_nda = np.around(weighted_nda_matrix / nda_norm_factor, 3)
```

```python
# normalized weighted grey sums of positive distances from the average (PDA)

norm_weighted_pda_str = get_str_matrix_of_grey_nums(norm_weighted_pda)
norm_weighted_pda_df = pd.DataFrame(norm_weighted_pda_str, columns = ['Candidate '+str(i+1) for i in range(num_of_candidates)])
norm_weighted_pda_df
```

```python
# normalized weighted grey sums of negative distances from the average (NDA)

norm_weighted_nda_str = get_str_matrix_of_grey_nums(norm_weighted_nda)
norm_weighted_nda_df = pd.DataFrame(norm_weighted_nda_str, columns = candidates)
norm_weighted_nda_df
```

```python
# calculating score

scores = []

for i in range(num_of_candidates):
    norm_weighted_pda_for_candidate = np.ravel(norm_weighted_pda[0,i,:])
    norm_weighted_nda_for_candidate = np.ravel(norm_weighted_nda[0,i,:])
    score = sum(norm_weighted_pda_for_candidate) + sum(norm_weighted_nda_for_candidate)
    score = np.round(score/4, 3)
    scores.append(score)
    
scores
```

```python
def get_ranks_for_candidates(scores):
    candidate_score_dict = {}
    c = 1
    for score in scores:
        candidate_score_dict[c] = score
        c += 1

    candidate_score_sorted = dict(sorted(candidate_score_dict.items(), key=lambda item: item[1]))
    ranks = [0 for i in range(num_of_candidates)]
    rank = 1
    for c in candidate_score_sorted:
        ranks[c-1] = rank
        rank += 1
        
    return ranks
```

```python
ranks = get_ranks_for_candidates(scores)
ranks
```

```python
scores_data = {'Candidate': candidates, 'Score': scores, 'Rank': ranks}
scores_df = pd.DataFrame(data=scores_data)
scores_df
```
