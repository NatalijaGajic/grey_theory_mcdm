```python
import pandas as pd
import numpy as np
```

```python
num_of_evaluators = 2
num_of_criteria = 15
criteria = ['C11','C21','C22','C23', 'C31', 'C32', 'C33', 'C34', 'C41', 'C42', 'C51', 'C52', 'C53',' C61', 'C62']
num_of_candidates = 10
```

```python
path_first_eval = '../g-edas-matrix-results-v.xlsx'
results_eval_frst_df = pd.read_excel(path_first_eval, header=0)
results_eval_frst_df = results_eval_frst_df.fillna(0)
results_eval_frst_df = results_eval_frst_df.drop(results_eval_frst_df.columns[[0]], axis=1)
results_eval_frst_df
```

```python
path_second_eval = '../g-edas-matrix-results-s.xlsx'
results_eval_scnd_df = pd.read_excel(path_second_eval, header=0)
results_eval_scnd_df = results_eval_scnd_df.fillna(0)
results_eval_scnd_df = results_eval_scnd_df.drop(results_eval_scnd_df.columns[[0]], axis=1)
results_eval_scnd_df
```

```python
ling_scale = ['Veoma lose', 'Lose', 'Srednje lose', 'Prosecno', 'Srednje dobro', 'Dobro', 'Veoma dobro']
grey_nums = [[0, 1], [1, 3], [3, 4], [4, 5], [5, 6], [6, 9], [9, 10]]
scale = {'Linguistic Scale': ling_scale, 'Value': [i for i in range(7)], 'Grey Numbers': grey_nums}
scale_df = pd.DataFrame(data=scale)
scale_df
```

```python
def print_grey_num(grey_num):
    return '[ ' + str(grey_num[0]) + ', ' + str(grey_num[1]) + ' ]'
                
def print_matrix(matrix):
    matrix_dim = matrix.shape[0]
    for i in range(matrix_dim):
        for j in range(matrix_dim):
            print(print_grey_num(matrix[i,j,:]), end =" ")
        print("\n")
        
def get_str_matrix_of_grey_nums(matrix):
    str_matrix = [[] for i in range(matrix.shape[0])]
    
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            grey_num = matrix[i,j]
            str_matrix[i].append(print_grey_num(grey_num))
            
    return str_matrix
```

```python
def get_grey_num_for_value(df, row_ind, col_ind):
    value = df.iloc[row_ind,col_ind]
    grey_num_for_value = grey_nums[value]
    return np.array(grey_num_for_value)
```

```python
# replace values with grey nums; get g-edas matrix

edas_matrix = np.zeros((num_of_candidates, num_of_criteria, 2), dtype=float)

for i in range(num_of_candidates):
    for j in range(num_of_criteria):
        frst_grey_num = get_grey_num_for_value(results_eval_frst_df, i, j)
        scnd_grey_num = get_grey_num_for_value(results_eval_scnd_df, i, j)
        edas_matrix[i,j] = (frst_grey_num + scnd_grey_num) / 2
        
edas_matrix = np.around(edas_matrix, 3)
```

```python
# initial decision-making matrix

str_matrix = get_str_matrix_of_grey_nums(edas_matrix)
str_matrix_df = pd.DataFrame(str_matrix, columns = criteria)
str_matrix_df
```
